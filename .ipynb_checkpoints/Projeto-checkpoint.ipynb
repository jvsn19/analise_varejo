```python
'''
    # Método para normalizar as coordenadas de um lugar
    # lat = 3.122231 lon = 5.522231 com precisão .2f normalizará para lat = 3.12 e lon 3.52
    # A partir do ajuste da precisão podemos definir locais próximos e, dessa forma, identificar,
    # por exemplo, aglomerado de lojas de um certo tipo
'''
def normalizeCoordinates(coordinates:pd.DataFrame, precision:int = 6):
    precision_str = '%.'+str(precision)+'f'
    float_str = lambda x : precision_str%x
    return coordinates.applymap(float_str)
```
